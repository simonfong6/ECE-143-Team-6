```python
from math import pi
def createRadar(player, df,Attributes):
    
    #Flatten values to list
    data = df.values.flatten().tolist()
    
    
    
    radius = len(Attributes)
    data += data [:1]
    
    angles = [n / radius * 2 * pi for n in range(radius)]
    angles += angles [:1]
    
    ax = plt.subplot(111, polar=True)

    plt.xticks(angles[:-1],Attributes)
    ax.plot(angles,data)
    ax.fill(angles, data, 'blue', alpha=0.1)
    
    
    ax.set_title(player)
    mpl
    dpi = 300
    mpl.rcParams['figure.dpi']= dpi
    plt.show()
```

```python
def var_count_radar(df,col,labels,df_row):
    '''
        Function will return a dataframe of the number of frequencies for
        'Always Offers to Pay',"Salty About Paying","Lets the Girl 
        Pay if She Wants".
        
        df: (DataFrame) dataframe
        
        Returns:
        
        df: dataframe with number of counts of each attribute in manners_eval
   '''
   
    
    my_dict = {}
    #print(labels)
    index = 0

    
    for name,values in df.iteritems():
        if name not in col:
            continue
        if index == len(labels):
            break
        name = labels[index]
        empty = []
        print(values)
        values = values.value_counts()[df_row]
        
        l = values.tolist()
       
        empty.append(l)
       
        my_dict[name] = empty
        index +=1
    
    print(my_dict)
    return pd.DataFrame(my_dict)
```

```python
salary_80k = filtered.loc[filtered['salary_over_80k']==True]
manners_eval = ['always_offer_to_pay','salty_about_paying',
                'let_the_girl_pay_if_she_wants']
labels =["Always Offers to Pay",
            "Salty About Paying",
            "Lets the Girl Pay if She Wants"]

salary_80k = var_count_radar(salary_80k,manners_eval,labels,True)

Attributes = ["Always Offers to Pay",
                  "Salty About Paying",
                  "Lets the Girl Pay if She Wants"]

createRadar("Salary Over 80k",salary_80k, Attributes)
```
